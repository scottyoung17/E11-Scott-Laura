Measured Standard Deviation & Theoretical Uncertainty Comparison

```python
#2a
a=df['Counts']
b=df1['Counts']
c=df2['Counts']
d=df3['Counts']
e=df4['Counts']
f=df5['Counts']
g=df6['Counts']
y=sum(a)
y1=sum(b)
y2=sum(c)
y3=sum(d)
y4=sum(e)
y5=sum(f)
y6=sum(g)
uncertainty=np.sqrt((np.sqrt(y))**2+(np.sqrt(y1))**2+(np.sqrt(y2))**2+(np.sqrt(y3))**2+(np.sqrt(y4))**2+(np.sqrt(y5))**2+(np.sqrt(y6))**2)
print ("Standard deviation for 1 cm tungsten shield =", std)
print ("Standard deviation for 2 cm tungsten shield =", std1)
print ("Standard deviation for 3 cm tungsten shield =", std2)
print ("Standard deviation for 2.12 mm aluminum shield =", std3)
print ("Standard deviation for 4.24 mm aluminum shield =", std4)
print ("Standard deviation for 6.36 mm aluminum shield =", std5)
print ("Standard deviation for 8.48 mm aluminum shield =", std6)
print("The total theoretical uncertainty is", uncertainty)
```

2b. What general trend do you see in the counts measured as a function of the thickness of
absorber material? Generally, as the thickness of the shield increases, the amount of counts measured decreases.

```python
#3
x=10
x1=20
x2=30
x3=2.12
x4=4.24
x5=6.36
x6=8.48
x_data=[x3,x4,x5,x6,x,x1,x2]
y_data=[y3,y4,y5,y6,y,y1,y2]
y_error=[np.sqrt(y3),np.sqrt(y4),np.sqrt(y5),np.sqrt(y6),np.sqrt(y),np.sqrt(y1),np.sqrt(y2)]
plt.plot(x_data,y_data)
plt.errorbar(x_data,y_data,yerr=y_error,fmt ='1')
plt.xlabel("Thickness (mm)")
plt.ylabel("Total Counts")
plt.show()
#We did not plot on a logarithmic scale because it did not make sense considering the max total amount
```
